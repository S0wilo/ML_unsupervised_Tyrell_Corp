```python
import numpy as np
import matplotlib.pyplot as plt
```

We define 2 normal distribution for height and weight of a person, with a total sample of 10000 and average height=1.75 and average weight=70, these are also the expected values of Z

```python
total_sample = 10000
height_mean = 1.75
height_std = 0.2
weight_mean = 70
weight_std = 10

height = np.random.normal(loc=height_mean, scale=height_std, size=(total_sample))
weight = np.random.normal(loc=weight_mean, scale=weight_std, size=(total_sample))


z = np.column_stack((height, weight))

expected_value_z = np.array([height_mean, weight_mean])
print(expected_value_z)
```

We plot a random sample of n element

```python
n = 100

selected_sample_idx = np.random.choice(z.shape[0], n, replace=False)
selected_sample = z[selected_sample_idx]
print(selected_sample.shape)

plt.scatter(selected_sample[:, 0], selected_sample[:, 1], alpha=0.5)
plt.title('Sampled Points from 2D Random Variable Z')
plt.xlabel('Height (X)')
plt.ylabel('Weight (Y)')
plt.grid(True)
plt.show()
```

We compute and plot the euclidean distance between the empirical average of the first n elements (with n from 1 to the total number of sample) and the expected value of Z

```python
euclidean_distances = []

for n in range(1, total_sample + 1):
        empirical_average = np.mean(z[:n, :], axis=0)
        euclidean_distance = np.linalg.norm(empirical_average - expected_value_z)
        euclidean_distances.append(euclidean_distance)

print(empirical_average)
        
plt.plot(range(1, total_sample + 1), euclidean_distances)
plt.title('Convergence of Empirical Average to Expected Value')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()
```
